# The great-circle arrangement

For each ray $r$, the directions **perpendicular** to $r$ form a great circle on the
sphere. That circle is where the deepest cut about $r$ — the one straight through the
centre — meets the surface. So drawing every ray's great circle shows the puzzle's
actual cut pattern, not just its axes.

Two facts worth watching for:

- **One circle per axis.** Opposite rays share a perpendicular plane, so `O · 6`
  (six rays) draws only **three** circles.
- **Regions are pieces.** The circles carve the sphere into regions, and those are
  the surface pieces of the *deepest-cut* puzzle on that ray-set. `O · 6` → 8 (the
  eight cubies of a 2x2x2); `I · 12` → 32 (the Pentultimate). The count is exact,
  from Euler's $V - E + F = 2$.

**Run all cells**, then use the dropdown and drag to rotate. Circles are coloured by
ray family.


```python
%matplotlib widget
import matplotlib.pyplot as plt
import ipywidgets as w
import great_circles as gc
import raysets

SETS = gc.the_16()

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection="3d")

def show(k=0):
    rs = SETS[k]
    circles, V, E, F = gc.arrangement(rs)
    ax.clear()
    gc.draw(ax, rs, split=False)          # whole circles: view will rotate
    ax.set_title(f"{raysets.label(rs)}     {circles} circles,  {F} regions")
    fig.canvas.draw_idle()

options = [(f"{raysets.label(rs)}   ({raysets.count(rs)} rays)", k)
           for k, rs in enumerate(SETS)]
w.interact(show, k=w.Dropdown(options=options, description="ray-set:"));
```

## Circles, vertices, and pieces

The **regions** column is the number of surface pieces you get by cutting each ray-set
as deep as possible — every cut passing through the centre. It is computed exactly:
count where the circles cross (vertices), count the arcs between crossings (edges),
and apply $V - E + F = 2$.


```python
print(f"{'ray-set':<16}{'rays':>5}{'circles':>9}{'vertices':>10}{'regions':>9}")
print("-" * 49)
for rs in SETS:
    circles, V, E, F = gc.arrangement(rs)
    print(f"{raysets.label(rs):<16}{raysets.count(rs):>5}{circles:>9}{V:>10}{F:>9}")
```

Two of these regions counts name a puzzle you can hold: **`O · 6` → 8** is the
2x2x2, and **`I · 12` → 32** is the Pentultimate (12 faces + 20 corners), the
deepest-cut Megaminx. The rest are the deepest-cut puzzles on their own ray-sets;
naming each one is a project in itself, and belongs with the cut-depth work in
[`../../3_cut_depths`](../../3_cut_depths) rather than here.

Note the arrangement only shows the *deepest* cut. Slide the cuts outward and the
piece structure changes — that whole ladder of depths is what folder 3 is about.
