# The 16 polyhedra — face-turning and vertex-turning

Every ray-set has a polyhedron attached to it, and in fact two, which are **duals**
of each other:

- **Face-turning** — put a plane perpendicular to each ray. Every ray becomes a
  *face*. `O · 6` is a cube, `I · 12` a dodecahedron: the Megaminx. Faces are
  coloured by ray family.
- **Vertex-turning** — take the convex hull of the ray tips. Every ray becomes a
  *vertex*. `O · 6` is now an octahedron, `I · 12` an icosahedron. Vertices carry
  the turning axes, so those are coloured by family; the faces go neutral.

It's the **same axis system** both ways — a vertex-turning octahedron and a
face-turning cube are the same puzzle geometry, `O · 6`. Duality just swaps which
features hold the axes.

**Run all cells**, then use the dropdown to pick a ray-set and the toggle to switch
face ↔ vertex. Drag to rotate.


```python
%matplotlib widget
import matplotlib.pyplot as plt
import ipywidgets as w
import polyhedra, raysets

SETS = polyhedra.the_16()

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection="3d")

def show(k=0, turning="face"):
    rs = SETS[k]
    ax.clear()
    if turning == "face":
        polyhedra.draw(ax, rs)
        feature = f"{raysets.count(rs)} faces"
    else:
        polyhedra.draw_dual(ax, rs)
        feature = f"{raysets.count(rs)} vertices"
    ax.set_title(f"{raysets.label(rs)}  ·  {turning}-turning  ·  {feature}")
    fig.canvas.draw_idle()

options = [(f"{raysets.label(rs)}   ({raysets.count(rs)} rays)", k)
           for k, rs in enumerate(SETS)]
w.interact(show,
           k=w.Dropdown(options=options, description="ray-set:"),
           turning=w.ToggleButtons(options=["face", "vertex"], description="turning:"));
```

## The whole list

Face count equals ray count for the face-turning solid; vertex count equals ray count
for its dual. Everything else swaps: the two columns below are the same numbers with
faces and vertices traded, and Euler's $V - E + F = 2$ holds for both.


```python
print(f"{'ray-set':<16}{'faces':>6}{'verts':>7}{'edges':>7}{'V-E+F':>7}   face-turning polyhedron")
print("-" * 78)

NAMED = {
    (4, 4, 6):        "tetrahedron  — Pyraminx",
    (6, 8, 12):       "cube  — 3x3x3",
    (8, 6, 12):       "octahedron",
    (12, 14, 24):     "rhombic dodecahedron",
    (12, 20, 30):     "dodecahedron  — Megaminx",
    (20, 12, 30):     "icosahedron",
    (30, 32, 60):     "rhombic triacontahedron",
    (14, 24, 36):     "truncated octahedron",
    (32, 60, 90):     "truncated icosahedron  — the soccer ball",
    (26, 48, 72):     "truncated cuboctahedron",
    (62, 120, 180):   "truncated icosidodecahedron",
}

for rs in SETS:
    faces, colours, V = polyhedra.polyhedron(rs)
    F, Vn = len(faces), len(V)
    E = sum(len(f) for f in faces) // 2
    name = NAMED.get((F, Vn, E), "")
    print(f"{raysets.label(rs):<16}{F:>6}{Vn:>7}{E:>7}{Vn - E + F:>7}   {name}")
```

The **vertex-turning** polyhedron of each row is the dual of the named one, so its
name is that solid's dual: the dual of `O · 6`'s cube is an **octahedron**, the dual
of `I · 12`'s dodecahedron is an **icosahedron**, and a tetrahedron is self-dual.

Two honest caveats on the names.

**They are combinatorial, not metric.** Every face plane sits at distance 1 from the
centre, which is not the spacing that makes an Archimedean solid's faces *regular*.
`O · 6+8`'s hexagons are hexagons in the right arrangement, but not regular ones. The
counts and structure match; the proportions do not.

**Five are unnamed on purpose.** `T · 4a+6`, `O · 6+12`, `O · 8+12`, `I · 12+30` and
`I · 20+30` don't match a solid that could be identified with confidence, so the
table leaves them blank rather than guessing. They are perfectly good convex
polyhedra — just not ones with famous names.
