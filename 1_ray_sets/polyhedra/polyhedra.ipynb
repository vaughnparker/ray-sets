# The 16 face-turning polyhedra

Every ray-set has a polyhedron attached to it, and the choice is forced rather than
invented. For a ray-set $R$, take the body

$$P = \{\, x : r \cdot x \le 1 \ \text{ for every ray } r \in R \,\}$$

Each ray's plane is tangent to the unit sphere exactly where that ray pierces it, so
**every ray contributes exactly one face** — and face-turning $P$ *is* turning about
the rays. `O · 6` gives a cube, `I · 12` a dodecahedron: the Megaminx.

**Run all cells**, then use the dropdown to step through the 16. Drag to rotate.
Faces are coloured by which ray family they came from.


```python
%matplotlib widget
import matplotlib.pyplot as plt
import ipywidgets as w
import polyhedra, raysets

SETS = polyhedra.the_16()
CACHE = {}                      # geometry is a second or two for the big ones

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection="3d")

def show(k=0):
    rs = SETS[k]
    if k not in CACHE:
        CACHE[k] = polyhedra.polyhedron(rs)
    faces, colours, V = CACHE[k]
    ax.clear()
    polyhedra.draw(ax, rs, faces, colours, V)
    ax.set_title(f"{raysets.label(rs)}     {len(faces)} faces")
    fig.canvas.draw_idle()

options = [(f"{raysets.label(rs)}   ({raysets.count(rs)} faces)", k)
           for k, rs in enumerate(SETS)]
w.interact(show, k=w.Dropdown(options=options, description="ray-set:"));
```

## The whole list

Face count always equals ray count, and Euler's $V - E + F = 2$ holds throughout —
so each of the 16 really is a convex polyhedron, not an accident of the construction.

Takes a few seconds: it solves every triple of face planes for all 16 bodies.


```python
print(f"{'ray-set':<16}{'faces':>6}{'verts':>7}{'edges':>7}{'V-E+F':>7}   polyhedron")
print("-" * 74)

NAMED = {
    (4, 4, 6):        "tetrahedron  — Pyraminx",
    (6, 8, 12):       "cube  — 3x3x3",
    (8, 6, 12):       "octahedron",
    (12, 14, 24):     "rhombic dodecahedron",
    (12, 20, 30):     "dodecahedron  — Megaminx",
    (20, 12, 30):     "icosahedron",
    (30, 32, 60):     "rhombic triacontahedron",
    (14, 24, 36):     "truncated octahedron",
    (32, 60, 90):     "truncated icosahedron  — the soccer ball",
    (26, 48, 72):     "truncated cuboctahedron",
    (62, 120, 180):   "truncated icosidodecahedron",
}

for rs in SETS:
    faces, colours, V = polyhedra.polyhedron(rs)
    F, Vn = len(faces), len(V)
    E = sum(len(f) for f in faces) // 2
    name = NAMED.get((F, Vn, E), "")
    print(f"{raysets.label(rs):<16}{F:>6}{Vn:>7}{E:>7}{Vn - E + F:>7}   {name}")
```

Two honest caveats on those names.

**They are combinatorial, not metric.** Every face plane here sits at distance 1
from the centre, which is not the spacing that makes an Archimedean solid's faces
*regular*. `O · 6+8`'s hexagons are genuine hexagons in the right arrangement, but
they are not regular hexagons. The counts and the structure match; the proportions
do not.

**Five are unnamed on purpose.** `T · 4a+6`, `O · 6+12`, `O · 8+12`, `I · 12+30`
and `I · 20+30` do not match a solid that could be identified with confidence, so
the table leaves them blank rather than guessing. They are perfectly good convex
polyhedra — they simply are not ones with famous names.
